```python
import pandas as pd
from sqlalchemy import create_engine

# 1. Connection Details (Matches your docker-compose.yaml)
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
engine.connect()

print("Connected to Postgres!")

# 2. Ingest Green Taxi Data (Parquet)
url_trips = "https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet"
print(f"Downloading {url_trips}...")

df_trips = pd.read_parquet(url_trips)

# Upload to SQL (chunking is not usually needed for parquet, but good practice)
print("Uploading trips to database...")
df_trips.to_sql(name='green_taxi_trips', con=engine, if_exists='replace', index=False)
print("Successfully uploaded Green Taxi Trips!")


# 3. Ingest Zones Data (CSV)
url_zones = "https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/taxi_zone_lookup.csv"
print(f"Downloading {url_zones}...")

df_zones = pd.read_csv(url_zones)

# Upload to SQL
print("Uploading zones to database...")
df_zones.to_sql(name='zones', con=engine, if_exists='replace', index=False)
print("Successfully uploaded Zones!")
```
